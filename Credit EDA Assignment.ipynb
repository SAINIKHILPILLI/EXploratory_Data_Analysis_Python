# 5.0 Top 10 correlations

```python
c=Target_difficulties[['AMT_GOODS_PRICE','AMT_INCOME_TOTAL','AMT_ANNUITY','DAYS_EMPLOYED',
  'DAYS_BIRTH', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH',
  'AMT_CREDIT']].corr(method = 'pearson')
c=c.where(np.triu(np.ones(c.shape),k=1).astype(np.bool))
df=c.unstack().reset_index()
df.columns = ['VARIABLE_1','VARIABLE_2','CORRELATION']
df.dropna(subset=['CORRELATION'],inplace=True)
df['CORRELATION_ABS']=df['CORRELATION'].abs()
df.sort_values('CORRELATION_ABS', ascending=False).head(15)
```

# 5.1  FILE_MERGING

```python
#Merging the two files 
combined_File = pd.merge(File1, File2, how='left', on=['SK_ID_CURR'])
```

```python
combined_File.head()
```

```python
#CHECKING THE WHOLE FILE
combined_File.describe()
```

```python
def uni_comb(x,y):
    sns.despine
    data = combined_File.pivot_table(values='SK_ID_CURR', 
                      index=x,
                      columns=y,
                      aggfunc='count')
    data=data.div(data.sum(axis=1),axis='rows')*100
    sns.set()
    data.plot(kind='bar',stacked=True,figsize=(20,10))
    plt.title(f'Effect Of {x} on Loan Approval')
    plt.xlabel(f'{x}')
    plt.ylabel(f'{y}%')
    plt.show()
```

```python
uni_comb('TARGET','NAME_CONTRACT_STATUS')
```

```python
uni_comb('CODE_GENDER','NAME_CONTRACT_STATUS')
```
